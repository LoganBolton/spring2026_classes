```python
import pandas as pd

file = "Asssignment4_Data.xlsx"
sheet = "Train"

df = pd.read_excel(file)
df = df.drop(columns=["House ID"])

target = "Construction type"
features = [col for col in df.columns if col != target]

# class priors P(Y)
print("Class priors:\n")
print(df[target].value_counts(normalize=True))
print()

# conditional distributions for each feature:
# mean and variance within each class
for feature in features:
    print("=" * 50)
    print("Feature:", feature)

    grouped = df.groupby(target)[feature]

    means = grouped.mean()
    variances = grouped.var(ddof=0)   # population variance for Gaussian NB

    print("\nMeans:")
    print(means)

    print("\nVariances:")
    print(variances)
    print()
```

I then computed the mean and variance of both local price and land area given each class. To do this, I used the following equations


The mean is

$$
\mu = \frac{1}{n}\sum_{i=1}^{n} x_i
$$

and the variance is

$$
\sigma^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i-\mu)^2
$$

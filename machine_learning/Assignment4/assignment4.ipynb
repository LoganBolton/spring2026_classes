```python

df = pd.read_excel("Asssignment4_Data.xlsx", sheet_name="Test")
df = df.drop(columns=["House ID"])
df = df.drop(columns=["Construction type"])

priors = {
    "Apartment": 0.35,
    "House": 0.35,
    "Condo": 0.30
}

means = {
    "Apartment": {
        "Local Price": 7.332743,
        "Bathrooms": 1.285714,
        "Land Area": 6.103857,
        "Living area": 1.505000,
        "# Garages": 1.214286,
        "# Rooms": 6.857143,
        "# Bedrooms": 3.428571,
        "Age of home": 38.714286
    },
    "House": {
        "Local Price": 5.760743,
        "Bathrooms": 1.071429,
        "Land Area": 6.630900,
        "Living area": 1.391714,
        "# Garages": 1.071429,
        "# Rooms": 6.142857,
        "# Bedrooms": 3.000000,
        "Age of home": 34.285714
    },
    "Condo": {
        "Local Price": 7.415900,
        "Bathrooms": 1.333333,
        "Land Area": 6.024667,
        "Living area": 1.553333,
        "# Garages": 1.333333,
        "# Rooms": 6.833333,
        "# Bedrooms": 3.333333,
        "Age of home": 39.666667
    }
}

vars = {
    "Apartment": {
        "Local Price": 11.207517,
        "Bathrooms": 0.275510,
        "Land Area": 9.101044,
        "Living area": 0.424941,
        "# Garages": 0.418367,
        "# Rooms": 1.551020,
        "# Bedrooms": 0.816327,
        "Age of home": 184.775510
    },
    "House": {
        "Local Price": 0.278604,
        "Bathrooms": 0.030612,
        "Land Area": 4.335326,
        "Living area": 0.038860,
        "# Garages": 0.602041,
        "# Rooms": 0.408163,
        "# Bedrooms": 0.285714,
        "Age of home": 138.775510
    },
    "Condo": {
        "Local Price": 17.719672,
        "Bathrooms": 0.305556,
        "Land Area": 5.396577,
        "Living area": 0.710620,
        "# Garages": 0.222222,
        "# Rooms": 2.138889,
        "# Bedrooms": 0.555556,
        "Age of home": 162.222222
    }
}

def gaussian(x, mean, var):
    top = math.exp(-((x - mean) ** 2) / (2 * var))
    bottom = math.sqrt(2 * math.pi * var)
    return top / bottom

for i in range(len(df)):
    row = df.iloc[i]

    apartment_prob = priors["Apartment"]
    house_prob = priors["House"]
    condo_prob = priors["Condo"]

    for col in df.columns:
        x = row[col]

        apartment_prob *= gaussian(x, means["Apartment"][col], vars["Apartment"][col])
        house_prob *= gaussian(x, means["House"][col], vars["House"][col])
        condo_prob *= gaussian(x, means["Condo"][col], vars["Condo"][col])

    total = apartment_prob + house_prob + condo_prob

    apartment_prob = apartment_prob / total
    house_prob = house_prob / total
    condo_prob = condo_prob / total

    if apartment_prob >= house_prob and apartment_prob >= condo_prob:
        prediction = "Apartment"
    elif house_prob >= apartment_prob and house_prob >= condo_prob:
        prediction = "House"
    else:
        prediction = "Condo"

    print("Example", i + 1)
    print("P(Apartment | x) =", round(apartment_prob, 6))
    print("P(House | x) =", round(house_prob, 6))
    print("P(Condo | x) =", round(condo_prob, 6))
    print("Predicted class =", prediction)
    print()
```

2. What is the effect of restricting the maximum depth of the tree? Try different depths and find the best value.

Increasing the depths of the tree increased the training accuracy but reducing the test accuracy. The best value I tried was with a depth of 2.

```python
best_depth = None
best_test_acc = 0

for depth in range(1, 11):
    clf = DecisionTreeClassifier(max_depth=depth, random_state=0)
    clf.fit(X_train, y_train)

    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)

    train_acc = accuracy_score(y_train, y_pred_train)
    test_acc = accuracy_score(y_test, y_pred_test)

    print("Max depth =", depth)
    print("Train Accuracy =", round(train_acc, 4))
    print("Test Accuracy =", round(test_acc, 4))
    print()

    if test_acc > best_test_acc:
        best_test_acc = test_acc
        best_depth = depth

print("Best depth =", best_depth)
print("Best test accuracy =", round(best_test_acc, 4))
```

```python
from sklearn.tree import DecisionTreeClassifier, _tree

clf = DecisionTreeClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)

feature_names = list(X_train.columns)
class_names = list(clf.classes_)
tree = clf.tree_

rules = []

def get_rules(node, conditions):
    if tree.feature[node] == _tree.TREE_UNDEFINED:
        class_index = tree.value[node][0].argmax()
        predicted_class = class_names[class_index]
        rules.append((conditions, predicted_class))
        return

    feature = feature_names[tree.feature[node]]
    threshold = tree.threshold[node]

    get_rules(tree.children_left[node], conditions + [(feature, "<=", threshold)])
    get_rules(tree.children_right[node], conditions + [(feature, ">", threshold)])

get_rules(0, [])

for i, rule in enumerate(rules):
    conditions = rule[0]
    predicted_class = rule[1]

    mask = pd.Series([True] * len(train_df))

    for feature, op, value in conditions:
        if op == "<=":
            mask = mask & (train_df[feature] <= value)
        else:
            mask = mask & (train_df[feature] > value)

    covered = train_df[mask]
    coverage_count = len(covered)
    coverage_fraction = coverage_count / len(train_df)

    correct = covered[covered["Construction type"] == predicted_class]
    if coverage_count > 0:
        accuracy = len(correct) / coverage_count
    else:
        accuracy = 0

    print("Rule", i + 1)
    print("IF ", end="")
    for j, (feature, op, value) in enumerate(conditions):
        if j > 0:
            print(" AND ", end="")
        print(f"{feature} {op} {value:.4f}", end="")
    print(f" THEN Construction type = {predicted_class}")
    print("Coverage count =", coverage_count)
    print("Coverage fraction =", round(coverage_fraction, 4))
    print("Accuracy =", round(accuracy, 4))
    print()
```
